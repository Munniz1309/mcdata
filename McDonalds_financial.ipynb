```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Importando os dados
df = pd.read_csv('/content/McDonalds_financial_statements.csv')
```

```python
# Visualizando os dados
df.head()
```

```python
# Visualizando as informações
df.info()
```

```python
X = df.drop('Year', axis=1).copy()
y = df.drop('Year', axis=1).copy()
```

```python
# Dividindo o conjunto de dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Criando o algoritmos de regressão
rf_model = RandomForestRegressor()

# 4. Treinando o Modelo
rf_model.fit(X_train, y_train)

# Fazendo as previsões
rf_predictions = rf_model.predict(X_test)
```

```python
# Criando o algoritmos de regressão
model_reg = LinearRegression()

# Treinando o modelo de regressão
model_reg.fit(X_train, y_train)

# Fazendo previsões
predictions_reg = model_reg.predict(X_test)
```

```python
# Métricas de acurácia RandomForestRegressor
mae = mean_absolute_error(y_test, rf_predictions)
mse = mean_squared_error(y_test, rf_predictions)
rmse = mean_squared_error(y_test, rf_predictions, squared=False)
r2 = r2_score(y_test, rf_predictions)

print(f'MAE: {mae}, MSE: {mse}, RMSE: {rmse}, R2: {r2}')
```

```python
# Métricas de acurácia LinearRegression
mae = mean_absolute_error(y_test, predictions_reg)
mse = mean_squared_error(y_test, predictions_reg)
rmse = mean_squared_error(y_test, predictions_reg, squared=False)
r2 = r2_score(y_test, predictions_reg)

print(f'MAE: {mae}, MSE: {mse}, RMSE: {rmse}, R2: {r2}')
```
